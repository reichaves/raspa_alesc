### Importa módulo requests

```python
import requests
```

### Importa módulo BeautifulSoup

```python
from bs4 import BeautifulSoup
```

### Guarda a url na variável url

```python
url = 'http://transparencia.alesc.sc.gov.br/gabinetes.php'
```

### Acessa a página, lê o conteúdo dela e salva na variável html

```python
html = requests.get('http://transparencia.alesc.sc.gov.br/gabinetes.php')
```

### Faz o parse e exibe o html

```python
soup = BeautifulSoup(html.content, 'html.parser')
```

```python
print(soup.prettify())
```
